```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python

#   for getting base url for different firms

url_b = 'https://www.moneycontrol.com/stocks/marketstats/bse-mostactive-stocks/sensex-4/'
r_b = requests.get(url_b)
soup_b = BeautifulSoup(r_b.content, 'html.parser')
firms = [i.text.strip() for i in soup_b.select('span.gld13.disin a') if i.text.strip() not in ('Watchlist', 'Portfolio')]
print(firms)
firmLink = [i['href'] for i in soup_b.select('span.gld13.disin a') if i.text.strip() not in ('Watchlist', 'Portfolio')]
firm_url = dict(zip(firms, firmLink))

```

```python
# generating news link for each firm

news = []
for firm, url in firm_url.items():
    r_n = requests.get(url)
    soup_n = BeautifulSoup(r_n.content, 'html.parser')
#     print(soup_n.select('.clearfix.viewmore.brdtp .FR a'))
    if soup_n.select_one('#news .tar a'):
        news_link = soup_n.select('#news .tar a.seemore')[-1]['href']
    else:
        news_link = soup_n.select_one('.clearfix.viewmore.brdtp .FR a')['href']
    
    news.append(news_link)
news
        
```

```python
# generating firms with link

firm_news_links = dict(zip(firms, news))
firm_news_links
```

```python
#  sample links, for a few firms(not used later)(redundant)

firms_news = {
    'ONGC' : 'https://www.moneycontrol.com/company-article/oilnaturalgascorporation/news/ONG',
    'Axis Bank' : 'https://www.moneycontrol.com/company-article/axisbank/news/AB16',
    'NTPC' : 'http://www.moneycontrol.com/company-article/ntpc/news/NTP',
    'HUL' : 'https://www.moneycontrol.com/company-article/hindustanunilever/news/HU',
    'Tata Steel' : 'http://www.moneycontrol.com/company-article/tatasteel/news/TIS',
    'Bharti AIrtel' : 'https://www.moneycontrol.com/company-article/bhartiairtel/news/BA08'

}
```

```python
# main code to generate articles and storing 

upperframe = [] 
frame = []
for name, url in firm_news_links.items():
#     print(name)

    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html.parser')
    home = 'https://www.moneycontrol.com'
    links=[home + i['href'] for i in soup.select('.rightCont .arial11_summ')]
    dates = [i.text.split('|')[1][:-1].strip() for i in soup.select('.a_10dgry')]    
   

#     print(links)
    for i,link in enumerate(links):
        print(link)
        
        firm = name
#         print('second')
        date = dates[i]
        print(date)
        
        r1 = requests.get(link)
        soup1 = BeautifulSoup(r1.content, 'html.parser')
        if soup1.select('.paywall') or soup1.select('.top_video_section'):
            continue
#         date = soup1.select_one('.article_schedule').text.split('/')[0]
        title = soup1.select_one('.article_title').text.strip()
        news = ''.join([i.text.strip() for i in soup1.select('.arti-flow p')])
        frame.append([firm,date, title, news])
#         time.sleep(2)
        
upperframe.extend(frame)
```

```python
# storing into pandas dataframe

data=pd.DataFrame(upperframe, columns=['Firm','Date', 'Title', 'Body'])
data
```

```python
# exporting to csv file

data.to_csv('news.csv', index=False)
```
